# 2. Pandas: Exploratory Data Analysis (EDA)

```python
#@title Do the necessary imports
import pandas as pd
```

```python
#@title Read the data (from a file)
# read the csv into pandas dataframe
df = pd.read_csv('titanic.csv')

display(df)
```

```python
#@title Read the data (from a github URL)
# https://github.com/datasciencedojo/datasets/blob/master/titanic.csv

raw_data = 'https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv'


# read the csv into pandas dataframe
df = pd.read_csv(raw_data)

# display the dataframe
display(df)
```

```python
#@title Getting the data type of the entire data
print("1. Data type of the data we read is: ")
print(type(df))
```

```python
#@title Getting the top 5 values
print("\n\n2. Here are your top 5 rows; You can also give an int to head()")
display(df.head())
```

```python
#@title Getting the top N values
print("\n\n2. Here are your top N rows; You can also give an int to head()")
display(df.head(9))
```

```python
#@title Getting the bottom 5 values
print("\n\n3. Here are your bottom 5 rows; You can also give an int to tail() ")
display(df.tail())
```

```python
#@title Getting the bottom N values
print("\n\n3. Here are your bottom N rows; You can also give an int to tail() ")
display(df.tail(3))
```

```python
#@title Basic queries:  shape, size and len

# Sample DataFrame
# data = {
#     'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
#     'Age': [25, 30, 35, 40, 45],
#     'Salary': [50000, 60000, 75000, 100000, 120000]
# }
# df = pd.DataFrame(data)

display(df)

# 1. Shape of the DataFrame (rows, columns)
print("Shape of the DataFrame (rows, columns):", df.shape)

# 2. Total size of the DataFrame (number of elements)
print("Total size of the DataFrame (number of elements):", df.size)

# 3. Number of rows in the DataFrame
print("Number of rows in the DataFrame:", len(df))
print("Number of rows in the DataFrame:", df.shape[0])
print("Number of columns in the DataFrame:", df.shape[1])
```

```python
#@title Count the rows (method 1)
# Same as "Finding the length of the data frame"
# Same as "how many rows are there in the data frame"
print("\n\n4. Here is the length (number of rows) of the dataframe ")
count = len(df)
print(count)
```

```python
#@title Count the rows (method 2)
# Same as "Finding the length of the data frame"
# Same as "how many rows are there in the data frame"
print("\n\n4. Here is the length (number of rows) of the dataframe ")
count = df.shape[0]
print(count)
```

```python
#@title Count the columns (method 1)
# Same as "Finding number of rows"
# Same as "how many rows are there in the data frame"
print("\n\n4. Count of columns in the dataframe: ")
count = len(df.columns)
print(count)
```

```python
#@title Count the columns (method 2)
# Same as "Finding number of rows"
# Same as "how many rows are there in the data frame"
print("\n\n4. Count of columns in the dataframe: ")
count = df.shape[1]
print(count)
```

```python
#@title finding the columns
print("\n\n5. Here are the column names of the dataframe; and the default datatype (Index) of the column list")
x = df.columns
print("Data type of df.columns: ", type(x))
print("Printing df.columns: ", x)

```

```python
#@title Finding the columns (as a list)
print("\n\n6. Here are the column names of the dataframe; and the datatype is changed to list type")
x= list(df.columns)
print("Data type of df.columns: ", type(x))
print("Printing df.columns: ", x)

print("Count of columns: ", len(x))

column_list = list(df.columns)
print(column_list)

for index, elem in enumerate(column_list):
  print(index, elem)
```

```python
#@title Finding the row (index) names
# TBD: How can we change the names?
print("\n\n7. Here are the row names of the dataframe; and the datatype (RangeIndex) of the row list.")
x = df.index
print(type(x))
print(x)
```

```python
#@title Finding the row (index) names as a list
print("\n\n8. Here are the row names of the dataframe; and the datatype is changed to list type")
x= list(x)
print(x)
print(type(x))
```

```python
#@title Getting the data types of the dataframe
print("\n\n10. data types in the data frame")
print(df.dtypes)
```

```python
#@title Finding the missing value counts
# NaN = Not a Number = It is Empty = Missing Value
nan_in_df = df.isnull().values.any()
print("\n\n12. Are there any missing values in the dataframe", nan_in_df)

#print(df.isnull())
df.isnull().sum()
```

```python
#@title Getting the info() of the dataframe
print("\n\n9 info summary about the data frame, this also lists the non-null count for each column")
display(df.info())
```

```python
#@title Basic Statistics (works only for the numerical columns)

# Sample DataFrame
# data = {
#     'Age': [25, 30, 35, 40, 45],
#     'Salary': [50000, 60000, 75000, 100000, 120000],
#     'Experience': [2, 4, 6, 8, 10]
# }
# df = pd.DataFrame(data)

# 1. Mean
print("Mean:\n", df.mean(), "\n")

# 2. Median
print("Median:\n", df.median(), "\n")

# 3. Mode
print("Mode:\n", df.mode(), "\n")

# 4. Standard Deviation
print("Standard Deviation:\n", df.std(), "\n")

# 5. Variance
print("Variance:\n", df.var(), "\n")

# 6. Minimum
print("Minimum:\n", df.min(), "\n")

# 7. Maximum
print("Maximum:\n", df.max(), "\n")
```

```python
#@title Descriptive Statistics for Numerical Columns
display(df)
df_stats = df.describe()
print("\n\n13. Descriptive Statistics of the dataframe (only numerical columns)\n", df_stats)
display('\n13. The type of df.describe(): ', type(df_stats))
```

```python
#@title Descriptive Statistics for Categorical Columns
df_stats = df.describe(include=['object', 'category'])
print("\n\n13. Descriptive Statistics of the dataframe (only categorical columns)\n", df_stats)
display('\n13. The type of df.describe(): ', type(df_stats))
```
