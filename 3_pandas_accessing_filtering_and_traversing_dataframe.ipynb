# loc vs iloc

```python
#@title Differences between loc vs iloc  (aka index location)

# Advantages and disadvantages
print("loc works well with meaninful lables and complex filtering")
print("loc does not work without labels")

print("1. Using loc to access data")
display(df.loc[1, "Name"])

# unlike range, loc function is grabbing the stopping index also
display(df.loc[0:2])

print("\n\n 2. Using iloc to access data")
# Advantage and disadvantage
print("iloc works well when labels are unknown or irrelevant")
print("You need to know exact index of rows and columns")
display(df.iloc[1, 3])

# iloc function honors the concept of range and does not grab the last index
display(df.iloc[0:2])


# loc[][] vs loc[a, b]
print("\n\nUsing loc[][] can be less effective and may lead to chained assignment error.")
print("Using loc[a, b] is more efficient and avoids errors. Use this method whenever possible.")
df[1:3]['Age'] = 5
```

#Traversing the dataframe (brute-force)

```python
#@title Traversing DataFrame rows

print("=====================\n")
print("\n\n DF Traversals: Row by Row")
print("Each ROW = Panda.Series Object")
print("=====================\n")

print("\n Remove the break to see the traversal of all rows")
print("\n\n1. DF Traversal: index method")
for index in df.index:  # for i in range(len(df)): will also work
    row_data = df.loc[index]
    # print("Whole Row = \n",row_data)
    # print("Row Type = ", type(row_data))
    name = df['Name'][index]  # Option 1
    name = df.loc[index, 'Name']  # Option 2
    name = df.iloc[index, 3] # Option 3

    age = df['Age'][index]
    age = df.loc[index, 'Age']

    ticket = df['Ticket'] [index]
    ticket = df.loc[index, 'Ticket']

    print(f"Index = {index} Name = {name} Age = {age} Ticket = {ticket}")
```

```python
#@title Traversing DataFrame rows

print("=====================\n")
print("\n\n DF Traversals: Row by Row")
print("Each ROW = Panda.Series Object")
print("=====================\n")

print("\n Remove the break to see the traversal of all rows")
print("\n\n1. DF Traversal: index method")
for index in df.index:  # for i in range(len(df)): will also work
    row_data = df.loc[index]
    print("Whole Row = \n",row_data)
    print("Row Type = ", type(row_data))
    name = df['Name'][index]  # Option 1
    name = df.loc[index, 'Name']  # Option 2
    name = df.iloc[index, 3] # Option 3

    age = df['Age'][index]
    age = df.loc[index, 'Age']

    ticket = df['Ticket'] [index]
    ticket = df.loc[index, 'Ticket']

    print(f"Index = {index} Name = {name} Age = {age} Ticket = {ticket}")
    break


print("\n\n2. DF Traversal: loc() method")
for index in range(len(df)):
    row_data = df.loc[index]  #
    print("Whole Row = \n",row_data)
    print("Row Type = ", type(row_data))
    name = df.loc[index, 'Name']  # selecting a single cell
    age = df.loc[index, 'Age'] # selecting a single cell
    ticket = df.loc[index, 'Ticket']  # selecting a single cell
    print(f"Index = {index} Name = {name} Age = {age} Ticket = {ticket}")
    break;

print("\n\n3. DF Traversal: iloc() method")
for index in range(len(df)):
    row_data = df.iloc[index]
    print("Whole Row = \n",row_data)
    print("Row Type = ", type(row_data))
    name = df.iloc[index, 3]
    age = df.iloc[index, 5]
    ticket = df.iloc[index, 8]
    print(f"Index = {index} Name = {name} Age = {age} Ticket = {ticket}")
    break;

'''iterrows() returns the iterator yielding each index value
   along with a series containing the data in each row.'''
print("\n\n4. DF Traversal: iterrows() method")
for index, row in df.iterrows():
    print("Whole Row = \n", row)
    print("Row Type = ", type(row))
    name = row['Name']
    age = row['Age']
    ticket = row['Ticket']
    print(f"Index = {index} Name = {name} Age = {age} Ticket = {ticket}")
    break;


'''itertuples() method will return an iterator yielding a named tuple for each row in the DataFrame.
   The first element of the tuple will be the row’s corresponding index value,
   while the remaining values are the row values.'''
print("\n\n5. DF Traversal: itertuples() method")
for row in df.itertuples():
    print("Whole Row = \n", row)
    print("Row Type = ", type(row))
    name = getattr(row, 'Name')
    age = getattr(row, 'Age')
    ticket = getattr(row, 'Ticket')
    print(f"Index = {index} Name = {name} Age = {age} Ticket = {ticket}")
    break;

```

```python
#@title Traversing the DataFrame Columns

# iteritems is deprecated in newer versions of pandas, use items() instead

# print("\n\n7. DF Traversal: iteritems() method")
# for column, content in df.iteritems( ):
#

print("=====================\n")
print("\n\n DF Traversals: Column by Column")
print("Each Column = Panda.Series Object")
print("=====================\n")


print("\n Remove the break to see the traversal of all columns")
print("\n\n6. DF Traversal: items() method")
for column, content in df.items( ):
    column_name = column
    column_content = content
    print("column name type " + str(type(column_name)))
    print("column content type " + str(type(column_content)))
    print(f'Column Name: {column_name}')
    print(f'Column Content: {column_content}', sep='\n')   # printing all content
    print(f'Column Content: {column_content[:5]}', sep='\n') # printing the first 5 rows of data
    break



print("\n\n8. DF Traversal: iterating over the columns")
for column in df:
    column_content = df[column]
    print("column name type " + str(type(column)))
    print("column content type " + str(type(column_content)))
    print(f'Column Name: {column_name}')
    print(f'Column Content: {column_content.values}', sep='\n') # printing all content
    print(f'Column Content: {column_content[:5]}', sep='\n') # printing the first 5 rows of data
    break
```

# Other ways of locating and filtering data

```python
#@title Boolean Indexing
# Example: Filter rows where passengers embarked at 'C'
df[df['Embarked'] == 'C']
```

```python
#@title Filtering: Single condition

# filtering
filter_condition = df['Age'] > 70
#display(df.loc[filter_condition])

# We can also create a new dataframe based on the filter
df2 = df.loc[filter_condition]
display(df2)

```

```python
#@title Filtering: multiple conditions
# Example: Passengers older than 60 and in Pclass 1

filtered_df = df[(df['Age'] > 60) & (df['Pclass'] == 1)]
display(filtered_df)
```

```python
#@title Filtering based on a dynamic .query( )
# Example: Filter rows using a query to get passengers in Pclass 1 and aged over 30
df.query('Pclass == 1 and Age > 30 and Survived == 0')
```

```python
#@title Filtering based on a custom function
# Sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 78]
}
df = pd.DataFrame(data)

# Custom function to filter rows
def filter_high_scores(row):
    return row['Score'] > 85

# Apply the function and filter rows
filtered_df = df[df.apply(filter_high_scores, axis=1)]

print(filtered_df)
```

```python
#@title Accessing with isin
# Example: Filter rows where Pclass is 1 or 3
df[df['Pclass'].isin([1, 3])]
```

```python
#@title Accessing some Random Rows
# Example: Select 5 random rows
df.sample(5)
df.sample(5, axis = 0)
```

```python
#@title Accessing some Random Columns
#Example: Select 5 random columns
df.sample(3, axis = 1)
```
