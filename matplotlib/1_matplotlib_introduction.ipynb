````python
#@title Two Variables: Basic Line Plot - plt.plot(x,y)

# Plotting Days vs Calories burned

# sample data
weeks = [x for x in range(1, 52)]
calories = [random.randint(200, 1000) for x in range(51)]

# Function to plot
plt.plot(weeks, calories)
#plt.plot(calories)  # x is optional and will default to 1 to len(y)

# function to show the plot
plt.show()
````

````python
#@title Setting the figure (plot size)
plt.figure(figsize=(18, 6))
# Function to plot
plt.plot(weeks, calories)
````

````python
#@title Setting the color, linewidth and linestyle explcitly
plt.figure(figsize=(12, 6))
plt.plot(weeks, calories, color='blue', linewidth=3, linestyle='dotted')
````

````python
#@title Setting the labels for x-axis, y-axis and the plot

plt.xlabel('Week')
plt.ylabel('Calories')
plt.title('My fitness tracker')

# Little trick to see the valid values for the line style; Make an intentional mistake
#plt.plot(days, calories, color='red', linewidth=5, linestyle='asf')

plt.plot(weeks, calories, color='red', linewidth=2, linestyle='dashed')
````

````python
#@title Formatting String   [[marker][line][color]
# Supported Colors: blue, green red, cyan, magenta, yellow, white, blac(k)
# Line Styles: solid (-) dashed(--) dash-dot (-.) dot(:)
# Markers: + x X * o (obvious ones)  8 (octagon), s (square) p(pentagon) P (plus filled)  h (hexagon) H (hexagon)  , (pixel)  D (diamond)
# plt.plot(days, calories, 'D-.m')


# explictly calling out marker,  linestyle and color
plt.plot(weeks, calories,  marker='x',  color='magenta', linestyle='-.')

# we can use formatting string = same effect as the above line
plt.plot(weeks, calories, '*-r')
````

````python
#@title Setting the marker size

plt.figure(figsize=(18, 6))

# Suppressign the line just to show the data points
plt.plot(weeks, calories, 'Dm')


# Suppressign the line just to show the data points; Set the size to the marker
plt.plot(weeks, calories, 'Dk', markersize = 10)


# Bring back the line
plt.plot(weeks, calories, 'D-k', markersize = 5)
````

````python
#@title Alpha property  (scale 0 to 1) to control the transparency (~ brightness)
# 0 makes it invisible 1; is the default; Anything in betweeen reduces the brightness

#plt.plot(weeks, calories, '<:g', linewidth = 2, markersize = 10, alpha = 0.4)

plt.plot(weeks, calories, 'o:k',   linewidth = 2,  alpha = 0.2)
````

````python
#@title Creating multiple data sets
days = [x for x in range(1, 31)]
calories_out = [random.randint(200, 1000) for x in range(30)]
calories_in = [random.randint(1500, 2500) for x in range(30)]
time = [random.randint(30, 90) for x in range(30)]

print("Days: ", days)
print("Calories Out: ", calories_out)
print("Calories In: ", calories_in)
print("Time: ", time)
````

````python
#@title set axes labels,  chart title, and legend
# https://matplotlib.org/stable/api/legend_api.html
plt.xlabel('Day')
plt.ylabel('Calories in/out, Distance')
plt.title('Fitness Tracker')

plt.plot(days, calories_in, 'o-r', label = 'Calories In')
plt.plot(days, calories_out, 'X-g', label = 'Calories Out')
plt.plot(days, time, 'x-b', label = 'Time')

# Legend Location best, upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center
plt.legend(loc='best')
#plt.legend(loc='upper left')
````

````python
#@title Controlling the font, size, style
# https://matplotlib.org/stable/gallery/text_labels_and_annotations/fonts_demo.html

import matplotlib.pyplot as plt

from matplotlib.font_manager import FontProperties

fig = plt.figure()
alignment = {'horizontalalignment': 'center', 'verticalalignment': 'baseline'}
yp = [0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2]
heading_font = FontProperties(size='large')

# Show family options
fig.text(0.1, 0.9, 'family', fontproperties=heading_font, **alignment)
families = ['serif', 'sans-serif', 'cursive', 'fantasy', 'monospace']
for k, family in enumerate(families):
    font = FontProperties(family=[family])
    fig.text(0.1, yp[k], family, fontproperties=font, **alignment)

# Show style options
styles = ['normal', 'italic', 'oblique']
fig.text(0.3, 0.9, 'style', fontproperties=heading_font, **alignment)
for k, style in enumerate(styles):
    font = FontProperties(family='sans-serif', style=style)
    fig.text(0.3, yp[k], style, fontproperties=font, **alignment)

# Show variant options
variants = ['normal', 'small-caps']
fig.text(0.5, 0.9, 'variant', fontproperties=heading_font, **alignment)
for k, variant in enumerate(variants):
    font = FontProperties(family='serif', variant=variant)
    fig.text(0.5, yp[k], variant, fontproperties=font, **alignment)

# Show weight options
weights = ['light', 'normal', 'medium', 'semibold', 'bold', 'heavy', 'black']
fig.text(0.7, 0.9, 'weight', fontproperties=heading_font, **alignment)
for k, weight in enumerate(weights):
    font = FontProperties(weight=weight)
    fig.text(0.7, yp[k], weight, fontproperties=font, **alignment)

# Show size options
sizes = [
    'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large']
fig.text(0.9, 0.9, 'size', fontproperties=heading_font, **alignment)
for k, size in enumerate(sizes):
    font = FontProperties(size=size)
    fig.text(0.9, yp[k], size, fontproperties=font, **alignment)

# Show bold italic
font = FontProperties(style='italic', weight='bold', size='x-small')
fig.text(0.3, 0.1, 'bold italic', fontproperties=font, **alignment)
font = FontProperties(style='italic', weight='bold', size='medium')
fig.text(0.3, 0.2, 'bold italic', fontproperties=font, **alignment)
font = FontProperties(style='italic', weight='bold', size='x-large')
fig.text(0.3, 0.3, 'bold italic', fontproperties=font, **alignment)

plt.xlabel('Day')
plt.ylabel('Calories in/out, Distance')
plt.title('Fitness Tracker')

plt.plot(days, calories_in, 'o-r', label = 'Calories In')
plt.plot(days, calories_out, 'X-g', label = 'Calories Out')
plt.plot(days, time, 'x-b', label = 'Time')

plt.grid()
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Random test data
np.random.seed(19680801)
all_data = [np.random.normal(0, std, size=100) for std in range(1, 4)]
labels = ['x1', 'x2', 'x3']

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

# rectangular box plot
bplot1 = ax1.boxplot(all_data,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax1.set_title('Rectangular box plot')

# notch shape box plot
bplot2 = ax2.boxplot(all_data,
                     notch=True,  # notch shape
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax2.set_title('Notched box plot')

# fill with colors
colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

# adding horizontal grid lines
for ax in [ax1, ax2]:
    ax.yaxis.grid(True)
    ax.set_xlabel('Three separate samples')
    ax.set_ylabel('Observed values')
````

````python
#@title Basic Box Plot
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(42)
data = [np.random.normal(0, std, 100) for std in range(1, 5)]

# Create a boxplot
plt.boxplot(data, patch_artist=True)

# Add labels and title
plt.title("Boxplot Example")
plt.xlabel("Data Groups")
plt.ylabel("Values")
plt.xticks([1, 2, 3, 4], ['Group 1', 'Group 2', 'Group 3', 'Group 4'])

# Show plot
plt.show()
````

````python
#@title Box Plot with custom colors
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)
fruit_weights = [
    np.random.normal(130, 10, size=100),
    np.random.normal(125, 20, size=100),
    np.random.normal(120, 30, size=100),
]
labels = ['peaches', 'oranges', 'tomatoes']
colors = ['peachpuff', 'orange', 'tomato']

fig, ax = plt.subplots()
ax.set_ylabel('fruit weight (g)')

bplot = ax.boxplot(fruit_weights,
                   patch_artist=True,  # fill with color
                   tick_labels=labels)  # will be used to label x-ticks

# fill with colors
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
````
