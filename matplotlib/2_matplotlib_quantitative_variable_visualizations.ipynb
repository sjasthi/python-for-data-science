# Visualizations for Quantitative Variables  (Discrete and Continuous)

```python
#@title Do the imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Discrete Variables

```python
#@title 1.1.Bar chart for Discrete Variables (Frequency Counts)
# Histogram for number of pets
number_of_pets = [0, 1, 2, 3, 4, 5]
counts = [30, 40, 20, 10, 5, 3]
plt.figure(figsize=(6, 4))
plt.bar(number_of_pets, counts, color='orange', width=0.6)
plt.title('Number of Pets Owned')
plt.xlabel('Number of Pets')
plt.ylabel('Count')
plt.show()
```

```python
#@title  1:2 Bar Chart for Discrete Variables (Frequency Bins)
# Example: Number of cars sold per price bracket
# Note: Price is continuous variable
#       However, by binning (putting values in buckets),
#       we are making this variable as discrete.
price_brackets = ['<$20k', '$20k-$30k', '$30k-$40k', '>$40k']
cars_sold = [50, 75, 60, 30]
plt.bar(price_brackets, cars_sold)
plt.title('Cars Sold Per Price Bracket')
plt.xlabel('Price Bracket')
plt.ylabel('Number of Cars Sold')
plt.show()
```

```python
#@title  2:1 Histogram for Discrete Variables
# Example: Distribution of exam scores
exam_scores = [55, 67, 67, 70, 80, 79, 85, 90, 90, 95, 100, 100]

#Note: By creating the bins, the discrete values are counted.
plt.hist(exam_scores, bins = 8, edgecolor = 'black')

plt.title('Distribution of Exam Scores')
plt.xlabel('Score Range')
plt.ylabel('Frequency')
plt.show()
```

```python
#@title 2.2.Histogram of discrete variables (using bins)
import matplotlib.pyplot as plt
import numpy as np

# Discrete data (number of pets per household)
data = [0, 1, 2, 1, 2, 3, 1, 2, 0, 1, 3, 2, 2, 1, 0, 4]

# Use bins that align with integers
bins = np.arange(min(data) - 0.5, max(data) + 1.5, 1)

plt.hist(data, bins=bins, edgecolor='black', align='mid', rwidth=0.8)
plt.xticks(np.arange(min(data), max(data)+1))  # Show discrete values on x-axis

plt.title('Number of Pets per Household')
plt.xlabel('Number of Pets')
plt.ylabel('Frequency')
plt.show()

```

```python
#@title  3: Dot Plot for Discrete Variables
# Example: Number of products sold per day
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
products_sold = [30, 40, 20, 50, 60]
plt.plot(days, products_sold, 'D')
plt.title('Products Sold Per Day')
plt.xlabel('Days')
plt.ylabel('Number of Products Sold')
plt.show()
```

```python
#@title 4.Frequency Table
# Frequency Table: Frequency of Dice Rolls
import matplotlib.pyplot as plt
import numpy as np

# Example data: Simulated dice rolls
np.random.seed(42)  # For reproducibility
dice_rolls = np.random.randint(1, 7, 100000)  # 1000 rolls of a 6-sided dice

# Calculate the frequency of each face
faces, counts = np.unique(dice_rolls, return_counts=True)

# Plot the frequency table as a bar chart
plt.bar(faces, counts, color='lightblue', edgecolor='black')
plt.xticks(faces)
plt.title('Frequency of Dice Rolls')
plt.xlabel('Dice Face')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#@title 5.Stacked Bar Chart
import matplotlib.pyplot as plt
import numpy as np

# Example data: Monthly sales segmented by product type
months = ['January', 'February', 'March', 'April', 'May', 'June']
product_types = ['Product A', 'Product B', 'Product C']

# Random sales data (in units)
np.random.seed(42)
sales_data = {
    product: np.random.randint(50, 200, size=len(months))
    for product in product_types
}

# Stacked bar chart
bar_width = 0.8
x = np.arange(len(months))

# Calculate cumulative sums for stacking
bottom_values = np.zeros(len(months))

plt.figure(figsize=(10, 6))

# Plot each product type
for product, sales in sales_data.items():
    plt.bar(x, sales, bar_width, label=product, bottom=bottom_values)
    bottom_values += sales

# Customization
plt.xticks(x, months)
plt.xlabel('Months')
plt.ylabel('Sales (units)')
plt.title('Monthly Sales Segmented by Product Type')
plt.legend(title='Product Types')
plt.tight_layout()

# Show the chart
plt.show()
```

```python
#@title  1.1: Histogram for Continuous Variables(Adding smooth curve)
# Using KDE (Kernel Density Estimate)
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Data
weights = np.random.normal(70, 10, 100)

# Histogram
plt.hist(weights, bins=10, edgecolor='black', alpha=0.5, density=True)

# KDE line
kde = gaussian_kde(weights)
x_vals = np.linspace(min(weights), max(weights), 200)
plt.plot(x_vals, kde(x_vals), color='red', label='KDE')

plt.title('Distribution of Weights with KDE')
plt.xlabel('Weight (kg)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
#@title 3.2. Box Plot with custom colors
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)
fruit_weights = [
    np.random.normal(130, 10, size=100),
    np.random.normal(125, 20, size=100),
    np.random.normal(120, 30, size=100),
]
labels = ['peaches', 'oranges', 'tomatoes']
colors = ['peachpuff', 'orange', 'tomato']

fig, ax = plt.subplots()
ax.set_ylabel('fruit weight (g)')

bplot = ax.boxplot(fruit_weights,
                   patch_artist=True,  # fill with color
                   tick_labels=labels)  # will be used to label x-ticks

# fill with colors
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

```python
#@title 4.Violin Plot
# Violin Plot: Age Distribution in a City
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Example data: Age distributions for different city zones
#np.random.seed(42)  # For reproducibility
zone_a = np.random.normal(30, 10, 200)  # Zone A: Mean 30, Std Dev 10
zone_b = np.random.normal(40, 12, 200)  # Zone B: Mean 40, Std Dev 12
zone_c = np.random.normal(35, 15, 200)  # Zone C: Mean 35, Std Dev 15

# Combine data into a single dataset
data = [zone_a, zone_b, zone_c]
zones = ['Zone A', 'Zone B', 'Zone C']

# Plot the violin plot
sns.violinplot(data=data)
plt.xticks(range(len(zones)), zones)
plt.title('Age Distribution in a City')
plt.xlabel('City Zones')
plt.ylabel('Age')
plt.show()
```

```python
#@title 9.Cumulative Frequency Plot (Ogive)
# example: Cumulative percentage of students scoring below certain marks
# Cumulative Frequency Plot (Ogive): Cumulative percentage of students scoring below certain marks
import matplotlib.pyplot as plt
import numpy as np

# Example data: Marks scored by students
marks = np.array([35, 50, 65, 70, 80, 85, 90, 95, 100])

# Generate cumulative frequencies
frequencies = np.array([5, 10, 20, 30, 40, 50, 60, 70, 80])

cumulative_frequencies = np.cumsum(frequencies)
cumulative_percentage = cumulative_frequencies / cumulative_frequencies[-1] * 100

# Plot the cumulative frequency (Ogive)
plt.plot(marks, cumulative_percentage, marker='o', linestyle='-')
plt.title('Cumulative Percentage of Students Scoring Below Certain Marks')
plt.xlabel('Marks')
plt.ylabel('Cumulative Percentage (%)')
plt.grid(True)
plt.show()
```

```python
#@title  11: Ridge Plot
# Example: Income distributions across different cities
from scipy.stats import gaussian_kde

# Simulate synthetic income data for different cities
np.random.seed(42)
cities = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"]
city_means = [70000, 65000, 60000, 58000, 55000]  # Average income per city
city_stddevs = [15000, 12000, 10000, 11000, 9000]  # Variation in income
income_data = {
    city: np.random.normal(mean, stddev, 1000)
    for city, mean, stddev in zip(cities, city_means, city_stddevs)
}

# Create a ridge plot for income distributions
plt.figure(figsize=(10, 8))
y_positions = range(len(cities))

for y, (city, incomes) in zip(y_positions, income_data.items()):
    # Compute KDE for income data
    kde = gaussian_kde(incomes)
    x = np.linspace(incomes.min() - 10000, incomes.max() + 10000, 500)
    kde_values = kde(x)

    # Normalize KDE values for better visualization
    kde_values = kde_values / kde_values.max()

    # Plot KDE curve
    plt.fill_between(x, y, y + kde_values, alpha=0.7, label=city)
    plt.plot(x, y + kde_values, color='black', linewidth=0.5)

# Customizing the plot
plt.yticks(y_positions, cities)
plt.xlabel("Income ($)")
plt.ylabel("City")
plt.title("Income Distributions Across Different Cities")
plt.tight_layout()
plt.show()
```
