#Time Series Data

```python
#@title Do the imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
#@title Introduction to Time Series Data
print("## Introduction to Time Series Data")
print("Time series data is a sequence of data points collected or recorded at specific time intervals.")
print("Examples include stock prices, weather records, and website traffic.")

# Generate a sample time series dataset
dates = pd.date_range(start="2023-01-01", end="2023-12-31", freq="D")
data = np.sin(np.linspace(0, 3 * np.pi, len(dates))) + np.random.normal(0, 0.2, len(dates))
time_series_df = pd.DataFrame({"Date": dates, "Value": data})

# Set the Date column as the index
time_series_df.set_index("Date", inplace=True)

display( time_series_df.head())

# Line Plot of Time Series Data
plt.figure(figsize=(12, 6))
plt.plot(time_series_df.index, time_series_df["Value"], label="Time Series Data", color="blue")
plt.title("Time Series Line Plot", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```

```python
#@title Rolling Mean Plot
rolling_mean = time_series_df["Value"].rolling(window=30).mean()
plt.figure(figsize=(12, 6))
plt.plot(time_series_df.index, time_series_df["Value"], label="Original Data", alpha=0.5)
plt.plot(time_series_df.index, rolling_mean, label="30-Day Rolling Mean", color="red", linewidth=2)
plt.title("Time Series with Rolling Mean", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```

```python
#@title Seasonal Decomposition
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(time_series_df["Value"], period=180, model="additive")
fig = decomposition.plot()
fig.set_size_inches(12, 10)
plt.suptitle("Seasonal Decomposition of Time Series", fontsize=16)
plt.show()
```

```python
#@title Scatter Plot with Time Series Data
plt.figure(figsize=(12, 6))
plt.scatter(time_series_df.index, time_series_df["Value"], label="Data Points", alpha=0.6, color="green")
plt.title("Scatter Plot of Time Series Data", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```

```python
#@title Histogram of Values
plt.figure(figsize=(12, 6))
time_series_df["Value"].hist(bins=30, color="purple", alpha=0.7)
plt.title("Histogram of Time Series Values", fontsize=16)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.grid()
plt.show()
```

```python
#@title Date Formatting Example
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(time_series_df.index, time_series_df["Value"], label="Time Series Data")
date_form = DateFormatter("%b-%Y")
ax.xaxis.set_major_formatter(date_form)
plt.title("Formatted Time Series Plot", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```

```python
#@title Summary
print("This notebook introduced time series data and demonstrated several visualization techniques:")
print("1. Line plot for observing trends.")
print("2. Rolling mean plot to smooth out fluctuations.")
print("3. Seasonal decomposition to identify components.")
print("4. Scatter plot for data point distribution.")
print("5. Histogram for value frequency analysis.")
print("6. Formatted time series plot with customized date formatting.")
```
