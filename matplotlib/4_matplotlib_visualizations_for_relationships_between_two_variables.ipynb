# Matplotlib Visualizations for Relationships Between Two Variables


```python
#@title Do the imports
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

## Section 1: Quantitative vs Quantitative


### 1. Scatter Plot
A scatter plot is used to visualize the relationship between two quantitative variables.

Example:
We'll use random data to simulate the relationship between height and weight.


```python
# Generate random data
# https://en.wikipedia.org/wiki/Scatter_plot
np.random.seed(42)
height = np.random.normal(170, 10, 100)
weight = np.random.normal(70, 15, 100)

height = [1,2,3,4,5]
weight = [10,18,32,44,49]


# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(height, weight, alpha=0.7, edgecolor='k')
plt.title('Scatter Plot of Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```


### 2. Line Chart
A line chart is used to visualize trends between two quantitative variables, especially over time.

Example:
We'll use a sine wave to simulate a time-dependent trend.


```python
# Generate time series data
time = np.linspace(0, 10, 100)
amplitude = np.sin(time)
#amplitude = np.cos(time)
#amplitude = np.tan(time)

# Line chart
plt.figure(figsize=(8, 6))
plt.plot(time, amplitude, color='blue', label='Sine Wave')
plt.title('Line Chart: Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```


### 3. Heatmap
A heatmap is used to visualize the density or strength of relationships between two quantitative variables.

Example:
We'll create a heatmap to display a correlation matrix for some generated data.


```python
import seaborn as sns

# Generate random data
data = np.random.rand(100, 3)
correlation_matrix = np.corrcoef(data, rowvar=False)

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', xticklabels=['Var1', 'Var2', 'Var3'], yticklabels=['Var1', 'Var2', 'Var3'])
plt.title('Heatmap of Correlation Matrix')
plt.show()
```


### 4. Hexbin Plot
A hexbin plot aggregates data points into hexagonal bins, useful for high-density scatter plots.

Example:
We'll visualize the relationship between two random variables.


```python
# Generate random data
x = np.random.randn(10000)
y = np.random.randn(10000)

# Hexbin plot
plt.figure(figsize=(8, 6))
plt.hexbin(x, y, gridsize=50, cmap='Blues')
plt.colorbar(label='Count')
plt.title('Hexbin Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```


### 5. Bubble Chart
A bubble chart is an extension of a scatter plot, where a third variable is represented by the size of the bubbles.

Example:
We'll simulate sales revenue (x), number of customers (y), and profit (bubble size).


```python
# Generate random data
np.random.seed(42)
sales = np.random.rand(50) * 100
customers = np.random.rand(50) * 50
profit = np.random.rand(50) * 500

# Bubble chart
plt.figure(figsize=(8, 6))
scatter = plt.scatter(sales, customers, s=profit, alpha=0.6, edgecolor='k', c=profit, cmap='viridis')
plt.title('Bubble Chart: Sales vs Customers with Profit Size')
plt.xlabel('Sales Revenue ($1000s)')
plt.ylabel('Number of Customers')
plt.colorbar(scatter, label='Profit ($)')
plt.grid(True)
plt.show()
```

## Section 2: Qualitative vs Quantitative


### 1. Box Plot
A box plot displays the spread and central tendency of a quantitative variable across qualitative categories.

Example:
We'll visualize test scores grouped by different classes.


```python
# Generate random data
categories = ['Class A', 'Class B', 'Class C']
scores = [np.random.normal(75, 10, 30), np.random.normal(85, 12, 30), np.random.normal(70, 15, 30)]

# Create a DataFrame
data = pd.DataFrame({'Category': np.repeat(categories, 30), 'Score': np.concatenate(scores)})

# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Score', data=data, palette='Set2')
plt.title('Box Plot of Scores by Class')
plt.xlabel('Class')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```


### 2. Violin Plot
A violin plot combines box plots and density plots to show data distribution and variation across categories.

Example:
We'll visualize the distribution of ages grouped by different job roles.


```python

# Generate random data
roles = ['Engineer', 'Manager', 'Analyst']
ages = [np.random.normal(30, 5, 50), np.random.normal(40, 7, 50), np.random.normal(35, 6, 50)]

# Create a DataFrame
data = pd.DataFrame({'Role': np.repeat(roles, 50), 'Age': np.concatenate(ages)})

# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='Role', y='Age', data=data, palette='muted', inner='quartile')
plt.title('Violin Plot of Age by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```


### 3. Bar Chart
A bar chart is used to compare the mean or total of a quantitative variable for each qualitative category.

Example:
We'll compare the average monthly sales across regions.


```python
# Generate random data
regions = ['North', 'South', 'East', 'West']
sales = [np.random.normal(5000, 800, 50), np.random.normal(6000, 900, 50), np.random.normal(4500, 700, 50), np.random.normal(5500, 750, 50)]

# Create a DataFrame
data = pd.DataFrame({'Region': np.repeat(regions, 50), 'Sales': np.concatenate(sales)})

# Bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Sales', data=data, estimator=np.mean, ci=None, palette='pastel')
plt.title('Average Monthly Sales by Region')
plt.xlabel('Region')
plt.ylabel('Average Sales ($)')
plt.grid(True)
plt.show()
```


### 4. Stacked Bar Chart
A stacked bar chart breaks down a quantitative variable across categories and subcategories.

Example:
We'll visualize monthly sales by region and product type.


```python
# Generate random data
regions = ['North', 'South', 'East', 'West']
product_types = ['Type A', 'Type B', 'Type C']
data = {
    'Region': np.repeat(regions, 3),
    'Product': product_types * 4,
    'Sales': np.random.randint(1000, 5000, size=12)
}
df = pd.DataFrame(data)

# Pivot table for stacked bar chart
pivot = df.pivot(index='Region', columns='Product', values='Sales')

# Stacked bar chart
pivot.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='viridis')
plt.title('Monthly Sales by Region and Product Type')
plt.xlabel('Region')
plt.ylabel('Sales ($)')
plt.legend(title='Product Type')
plt.grid(True)
plt.show()
```

## Section 3: Qualitative vs Qualitative


### 1. Mosaic Plot
A mosaic plot visualizes the relationship between two categorical variables using proportional rectangles.

Example:
We'll visualize the relationship between gender and preferred product categories.


```python
from statsmodels.graphics.mosaicplot import mosaic

# Example DataFrame
df = pd.DataFrame({
    'Gender': ['Male', 'Male', 'Female', 'Female'],
    'Product': ['A', 'B', 'A', 'B'],
    'Count': [10, 20, 15, 25]
})

# Convert to flat dictionary
mosaic_data = df.set_index(['Gender', 'Product'])['Count'].to_dict()

# Mosaic plot
plt.figure(figsize=(8, 6))
mosaic(mosaic_data, title='Mosaic Plot: Gender vs Product Preference', gap=0.02)
plt.show()

```


### 2. Stacked Bar Chart
A stacked bar chart shows proportions of subcategories within each category.

Example:
We'll visualize the distribution of voting preferences by age group.


```python
# Generate random data
age_groups = ['18-25', '26-35', '36-50', '50+']
parties = ['Party A', 'Party B', 'Party C']
data = {
    'Age Group': np.repeat(age_groups, 3),
    'Party': parties * 4,
    'Votes': np.random.randint(10, 100, size=12)
}
df = pd.DataFrame(data)

# Pivot table for stacked bar chart
pivot = df.pivot(index='Age Group', columns='Party', values='Votes')

# Stacked bar chart
pivot.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='tab10')
plt.title('Voting Preferences by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Votes')
plt.legend(title='Party')
plt.grid(True)
plt.show()
```


### 3. Grouped Bar Chart
A grouped bar chart places subcategories side by side for better comparison.

Example:
We'll compare the distribution of meal preferences (vegetarian, non-vegetarian) across genders.


```python
# Generate random data
genders = ['Male', 'Female']
meal_types = ['Vegetarian', 'Non-Vegetarian']
data = {
    'Gender': np.repeat(genders, 2),
    'Meal Type': meal_types * 2,
    'Count': np.random.randint(20, 50, size=4)
}
df = pd.DataFrame(data)

# Grouped bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Count', hue='Meal Type', data=df, palette='husl')
plt.title('Meal Preferences by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Meal Type')
plt.grid(True)
plt.show()
```


### 4. Heatmap
A heatmap uses color gradients to visualize the frequency of combinations of two categorical variables.

Example:
We'll display the frequency of combinations of education level and income bracket.


```python
# Generate random data
education_levels = ['High School', 'Bachelor', 'Master', 'PhD']
income_brackets = ['<50K', '50-100K', '>100K']
data = np.random.randint(10, 100, size=(len(education_levels), len(income_brackets)))

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, fmt='d', cmap='YlGnBu', xticklabels=income_brackets, yticklabels=education_levels)
plt.title('Heatmap: Education Level vs Income Bracket')
plt.xlabel('Income Bracket')
plt.ylabel('Education Level')
plt.show()
```
