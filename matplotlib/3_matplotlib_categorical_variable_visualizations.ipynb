````python
#@title Bar Chart (Sorted)
import matplotlib.pyplot as plt

# Original data
colors = ['Red', 'Blue', 'Green', 'Yellow', 'Purple']
counts = [15, 20, 35, 25, 10]

# Sort data by count
sorted_data = sorted(zip(counts, colors), reverse=True)  # Sort descending
sorted_counts, sorted_colors = zip(*sorted_data)

# Optional: map sorted colors to their actual color values
color_map = {'Red': 'red', 'Blue': 'blue', 'Green': 'green', 'Yellow': 'yellow', 'Purple': 'purple'}
sorted_color_values = [color_map[color] for color in sorted_colors]

# Plot the sorted bar chart
plt.figure(figsize=(6, 4))
plt.bar(sorted_colors, sorted_counts, color=sorted_color_values)
plt.title('Favorite Colors (Sorted)')
plt.xlabel('Colors')
plt.ylabel('Count')
plt.show()
````

````python
#@title Pie Chart
# Pie chart for favorite colors
colors = ['Red', 'Blue', 'Green', 'Yellow', 'Purple']

counts = [15, 20, 60, 25, 10]
plt.figure(figsize=(6, 6))
plt.pie(counts, labels = colors, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'yellow', 'purple'])
plt.title('Favorite Colors (Pie Chart)')
plt.show()
````

````python
#@title Stacked bar chart
#for favorite colors by gender
counts_men = [18, 12, 15, 10, 5]
counts_women = [7, 8, 20, 15, 5]
colors = ['Red', 'Blue', 'Green', 'Yellow', 'Purple']

x = np.arange(5)
#print(x)
width = 0.6

plt.figure(figsize=(8, 5))
plt.bar(x, counts_men, width, label='Men', color='lightblue')
plt.bar(x, counts_women, width, bottom=counts_men, label='Women', color='pink')
plt.xticks(x, colors)
plt.title('Favorite Colors by Gender (Stacked Bar Chart)')
plt.xlabel('Colors')
plt.ylabel('Count')
plt.legend()
plt.show()
````

````python
#@title Grouped Bar Chart

# Data
colors = ['Red', 'Blue', 'Green', 'Yellow', 'Purple']
children_counts = [12, 15, 8, 10, 5]
adults_counts = [8, 12, 20, 15, 10]

# X-axis positions
x = np.arange(len(colors))  # Array of indices for the categories
width = 0.4  # Width of each bar

# Plot grouped bar chart
plt.figure(figsize=(12, 5))
plt.bar(x - width/2, children_counts, width, label='Children', color='skyblue')
plt.bar(x + width/2, adults_counts, width, label='Adults', color='orange')

# Add labels and title
plt.xlabel('Colors')
plt.ylabel('Count')
plt.title('Favorite Colors by Age Group')
plt.xticks(x, colors)  # Replace x-axis indices with category names
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
````

````python
#@title Word Cloud
from wordcloud import WordCloud

# Data: Words and their frequencies
word_frequencies = {
    'Python': 100,
    'Java': 35,
    'C++': 25,
    'JavaScript': 30,
    'Ruby': 15,
    'HTML': 20,
    'CSS': 20,
    'SQL': 30,
    'Swift': 10,
    'PHP': 10,
    'R': 10,
    'Go': 10,
    'MATLAB': 10,
    'Scala': 10,
}

# Generate the Word Cloud
wordcloud = WordCloud(
    width = 800,
    height = 400,
    background_color='white',
    colormap='autumn'
).generate_from_frequencies(word_frequencies)

# Display the Word Cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Remove axes
plt.title('Word Cloud of Programming Languages', fontsize=16)
plt.show()
````

````python
#@title Mosaic Plot
from statsmodels.graphics.mosaicplot import mosaic

# Data: Drink preferences by gender
data = {
    ('Men', 'Coffee'): 40,
    ('Men', 'Tea'): 20,
    ('Men', 'Juice'): 15,
    ('Women', 'Coffee'): 30,
    ('Women', 'Tea'): 35,
    ('Women', 'Juice'): 25,
}

# Create Mosaic Plot
plt.figure(figsize=(8, 6))
mosaic(data, labelizer=lambda key: f'{key[1]} ({key[0]})', title='Drink Preferences by Gender')
plt.title('Mosaic Plot: Drink Preferences by Gender', fontsize=14)
plt.show()
````

````python
#@title Bubble Chart
# Data
products = ['A', 'B', 'C', 'D', 'E']
prices = [16, 15, 8, 12, 20]  # Product prices
units_sold = [500, 80, 150, 600, 500]  # Units sold
revenues = [p * u for p, u in zip(prices, units_sold)]  # Total revenue
print(revenues)


# Bubble Chart
plt.figure(figsize=(8, 6))
# Scale bubble sizes for better visualization
bubble_sizes = [r / 10 for r in revenues]
plt.scatter(prices, units_sold, s=bubble_sizes, alpha=0.6, c=range(len(products)), cmap='viridis', edgecolors='black')

# Annotate bubbles with product names
for i, product in enumerate(products):
    plt.text(prices[i], units_sold[i], product, fontsize=10, ha='center', va='center')

# Add titles and labels
plt.title('Bubble Chart: Product Sales Performance', fontsize=14)
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.colorbar(label='Product Index')
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.show()
````

````python
#@title Sunburst Chart
import plotly.graph_objects as go

# Data for the Sunburst Chart
labels = [
    "World",             # Level 0
    "North America",     # Level 1
    "Europe",            # Level 1
    "USA",               # Level 2
    "Canada",            # Level 2
    "Germany",           # Level 2
    "France",            # Level 2
    "Electronics",       # Level 3
    "Furniture",         # Level 3
    "Clothing",          # Level 3
    "Electronics",       # Level 3
    "Furniture",         # Level 3
]
parents = [
    "",                  # "World" has no parent
    "World",             # "North America" belongs to "World"
    "World",             # "Europe" belongs to "World"
    "North America",     # "USA" belongs to "North America"
    "North America",     # "Canada" belongs to "North America"
    "Europe",            # "Germany" belongs to "Europe"
    "Europe",            # "France" belongs to "Europe"
    "USA",               # "Electronics" belongs to "USA"
    "USA",               # "Furniture" belongs to "USA"
    "Canada",            # "Clothing" belongs to "Canada"
    "Germany",           # "Electronics" belongs to "Germany"
    "France",            # "Furniture" belongs to "France"
]
values = [
    100,                 # "World" total
    60,                  # "North America" total
    40,                  # "Europe" total
    30,                  # "USA" total
    30,                  # "Canada" total
    20,                  # "Germany" total
    20,                  # "France" total
    15,                  # Electronics in "USA"
    15,                  # Furniture in "USA"
    30,                  # Clothing in "Canada"
    10,                  # Electronics in "Germany"
    10,                  # Furniture in "France"
]

# Create Sunburst Chart
fig = go.Figure(go.Sunburst(
    labels=labels,
    parents=parents,
    values=values,
    branchvalues="total",  # Distribute values proportionally to children
))

# Add title and show
fig.update_layout(
    title="Sunburst Chart: Hierarchical Sales Data",
    margin=dict(t=30, l=0, r=0, b=0)
)
fig.show()
````

````python
#@title Pareto Chart
import matplotlib.pyplot as plt
import numpy as np

# Data: Defect categories and their frequencies
categories = ['Scratches', 'Dents', 'Cracks', 'Color Issues', 'Other']
frequencies = [50, 30, 20, 10, 5]

# Sort categories and frequencies in descending order
sorted_indices = np.argsort(frequencies)[::-1]
categories = np.array(categories)[sorted_indices]
frequencies = np.array(frequencies)[sorted_indices]

# Calculate cumulative percentage
cumulative = np.cumsum(frequencies) / np.sum(frequencies) * 100

# Plot Pareto Chart
fig, ax1 = plt.subplots(figsize=(8, 5))

# Bar chart for defect frequencies
ax1.bar(categories, frequencies, color='skyblue', alpha=0.7)
ax1.set_xlabel('Defect Categories', fontsize=12)
ax1.set_ylabel('Frequency', fontsize=12, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_title('Pareto Chart: Defect Analysis', fontsize=14)

# Line chart for cumulative percentage
ax2 = ax1.twinx()
ax2.plot(categories, cumulative, color='red', marker='o', linestyle='-', linewidth=2)
ax2.set_ylabel('Cumulative Percentage', fontsize=12, color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.axhline(y=80, color='green', linestyle='--', linewidth=1)  # Highlight 80% line

# Annotate cumulative percentages
for i, val in enumerate(cumulative):
    ax2.text(i, val + 2, f'{val:.1f}%', ha='center', fontsize=10, color='red')

plt.tight_layout()
plt.show()
````

````python
#@title Heatmap
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Data: Satisfaction levels for three products
data = np.array([
    [50, 30, 20],  # Product A
    [40, 35, 25],  # Product B
    [60, 25, 15],  # Product C
])
categories = ['Satisfied', 'Neutral', 'Unsatisfied']
products = ['Product A', 'Product B', 'Product C']

# Create heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(data, annot=True, fmt="d", cmap="YlGnBu", xticklabels=categories, yticklabels=products)
plt.title("Customer Satisfaction Heatmap")
plt.xlabel("Satisfaction Level")
plt.ylabel("Products")
plt.show()
````

````python
#@title Sankey Diagram
from matplotlib.sankey import Sankey

# Data: Sankey example
sankey = Sankey(unit=None)
sankey.add( flows=[100, -60, -30, -10],
            labels=['Enrolled', 'Graduated', 'Dropped Out', 'Still Enrolled'], orientations=[0, 1, 1, -1])
sankey.finish()
plt.title("Student Transitions Sankey Diagram")
plt.show()
````

````python
#@title Radar (Spider) Chart
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Communication', 'Coding', 'Leadership', 'Problem Solving', 'Time Management']
candidate1 = [8, 7, 6, 8, 9]
candidate2 = [7, 9, 8, 6, 7]

# Setup for radar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]  # Complete the circle

# Add data
candidate1 += candidate1[:1]
candidate2 += candidate2[:1]

# Create radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, candidate1, color='blue', alpha=0.25, label='Candidate 1')
ax.fill(angles, candidate2, color='orange', alpha=0.25, label='Candidate 2')
ax.set_yticks(range(1, 11, 2))
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.legend()
plt.title('Skill Comparison Radar Chart')
plt.show()
````

````python
#@title Network Graph
import networkx as nx

# Create graph
G = nx.Graph()

# Add nodes and edges
G.add_edges_from([
    ('Alice', 'Bob'),
    ('Alice', 'Charlie'),
    ('Bob', 'Diana'),
    ('Charlie', 'Diana'),
    ('Diana', 'Eve'),
    ('Eve', 'Alice')
])

# Draw the graph
plt.figure(figsize=(6, 6))
nx.draw_networkx(G, with_labels=True, node_color='skyblue', edge_color='gray', node_size=1000, font_size=10)
plt.title("Social Network Graph")
plt.show()
````
