```python
#@title Adding / Recoding a column
# (create new column based on other column values)
# Southampton, Cherbourg, and Queenstown

display(df.head())

def map_names(row):
    name = row['Embarked']
    if name == 'S':
        return 'Southampton'
    if name == 'C':
        return "Cherbourg"
    if name == 'Q':
        return "Queenstown"

    return "Unknown"

def map_age_to_bins(row):
  age_bin = 'Unknown'
  age = row['Age']
  if age < 13:
    age_bin = 'Child'
  elif (age >= 13) and (age < 20):
    age_bin = 'Teenager'
  elif (age>=20):
    age_bin = 'Adult'

  return age_bin

# we now want to call the above function while recoding the column "length" to create a new column "length_str"
print(' \n\n Recoding the length column to length_str based on a custom function' )
df['Ebmarked_FullName'] = df.apply(lambda row: map_names(row), axis=1)
#df.drop('Embarked', axis=1, inplace=True)  #if we want to drop the old column
display(df)

print(' \n\n Recoding the age column to age_bin based on a custom function' )
df['age_bin'] = df.apply(lambda row: map_age_to_bins(row), axis=1)
#df.drop('Embarked', axis=1, inplace=True)  #if we want to drop the old column
display(df.head())
```

```python
# Recoding of the columns
# Using the custom function
# Let us say we want to create a new column called length_str which qualifies the length of the string
# short <= 30
# one_hour (31 go 60)
# two_hours (61 to 120)
# three_hours (121 to 180)
# very long (> 181)




def length2str(row):
    length = row['length']
    if length <= 30:
        return "short"
    if length < 60:
        return "one_hour"
    if length < 120:
        return "two_hours"
    if length < 180:
        return "three_hours"
    return "very_long"

# we now want to call the above function while recoding the column "length" to create a new column "length_str"
print(' \n\n Recoding the length column to length_str based on a custom function' )
movies_df['length_str'] = movies_df.apply(lambda row: length2str(row), axis=1)
#movies_df.drop('length', axis=1, inplace=True)  #if we want to drop the old column
display(movies_df)
```
